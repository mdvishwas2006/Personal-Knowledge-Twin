```python
import faiss
import numpy as np

dimension = embeddings.shape[1]

index = faiss.IndexFlatL2(dimension)

index.add(np.array(embeddings))

print("Knowledge Base Created")
print("Total Chunks Stored:", index.ntotal)
```

```python
def search(query, query_expansion=True, k=5):
    if query_expansion:
        # Use Gemini to expand the query
        expansion_prompt = f"Generate 3-5 alternative phrasing or related keywords for the following query, separated by commas. Do not include introductory phrases. Query: {query}"
        expanded_text = model_gemini.generate_content(expansion_prompt).text
        expanded_queries = [q.strip() for q in expanded_text.split(',') if q.strip()]
        # Include the original query and expanded queries
        all_queries = [query] + expanded_queries
        query_embeddings = model.encode(all_queries)
        # Average the embeddings to get a combined query embedding
        query_embedding = np.mean(query_embeddings, axis=0, keepdims=True)
    else:
        query_embedding = model.encode([query])

    distances, indices = index.search(
        np.array(query_embedding),
        k=k
    )

    results = []

    for idx in indices[0]:
        results.append(chunks[idx])

    return results
```

### Ask Multiple Questions to PDF Knowledge Base

To optimize API usage and reduce the number of `generate_content` calls, you can combine multiple questions into a single prompt. The following function `ask_pdf_multi_questions` takes a list of questions and formats them into one prompt for the model to answer.

```python
def ask_pdf_multi_questions(questions):
    knowledge = open("knowledge.txt", "r", encoding="utf-8").read()

    # Format multiple questions into a single string for the prompt
    formatted_questions = "\n".join([f"{i + 1}. {q}" for i, q in enumerate(questions)])

    prompt = f"""
    Answer the following questions only using the information below.

    INFORMATION:
    {knowledge}

    QUESTIONS:
    {formatted_questions}

    ANSWERS:
    """

    response = model_gemini.generate_content(prompt)
    return response.text
```

### Demonstration of `ask_pdf_multi_questions`

```python
questions_to_ask = [
    "What are the different types of database languages?",
    "Explain the Hierarchical Data Model.",
    "What are the advantages of a Relational Data Model?"
]

combined_answers = ask_pdf_multi_questions(questions_to_ask)
print(combined_answers)
```

### Batch Processing More Questions

To demonstrate the batch processing capability of `ask_pdf_multi_questions`, here's an example with an expanded list of questions.

```python
extended_questions_to_ask = [
    "What are the different types of database languages?",
    "Explain the Hierarchical Data Model.",
    "What are the advantages of a Relational Data Model?",
    "What is DDL used for?",
    "List some DML commands.",
    "What is the primary function of DQL?",
    "Give an example of a DCL command.",
    "What does TCL manage in DBMS?"
]

combined_answers_extended = ask_pdf_multi_questions(extended_questions_to_ask)
print(combined_answers_extended)
```

```python
import faiss
import numpy as np
from sentence_transformers import SentenceTransformer
```

```python
with open("knowledge.txt", "r", encoding="utf-8") as f:
    text = f.read()

print(text[:500])
```

```python
def split_text(text, chunk_size=500):
    chunks = []

    for i in range(0, len(text), chunk_size):
        chunks.append(text[i:i+chunk_size])

    return chunks

chunks = split_text(text)

print("Number of chunks:", len(chunks))
print(chunks[0])
```

```python
embedding_model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
embeddings = embedding_model.encode(chunks)

print(embeddings.shape)
```

```python
dimension = embeddings.shape[1]

index = faiss.IndexFlatL2(dimension)

index.add(np.array(embeddings))

print("FAISS Index Created")
```

```python
def retrieve(query, k=3):

    query_embedding = embedding_model.encode([query])

    distances, indices = index.search(
        np.array(query_embedding),
        k
    )

    results = []

    for idx in indices[0]:
        results.append(chunks[idx])

    return "\n".join(results)
```

```python
print(retrieve("What is this document about?"))
```

```python
def ask_rag(question):

    context = retrieve(question)

    prompt = f"""
    Answer the question using only the context below.

    Context:
    {context}

    Question:
    {question}
    """

    response = model_gemini.generate_content(prompt)

    return response.text
```

```python
print(ask_rag("Summarize this document"))
```
